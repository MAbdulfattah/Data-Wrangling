```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from datetime import datetime
pd.options.display.max_rows = 100
```

```python
final_df = pd.read_csv('full_dataset.csv', index_col=False)
final_df = final_df.drop(['Unnamed: 0'], axis=1)

weather_df = pd.read_csv('5y_weather_forcasts.csv')
weather_df = weather_df.drop(['Unnamed: 0'], axis=1)
```

# Feature Extraction

```python
final_df.Date = pd.to_datetime(final_df['Date'], format='%m/%d/%Y')
final_df['month'] = [int(x.month) for x in final_df.Date]

winter = final_df[final_df['month'].isin([12, 1, 2])]
spring = final_df[final_df['month'].isin([3, 4, 5])]
summer = final_df[final_df['month'].isin([6, 7, 8])]
fall = final_df[final_df['month'].isin([9, 10, 11])]
```

```python
def make_season(x, numeric=False):
    if x in [12, 1, 2]:
        return 0 if numeric else 'winter'
    if x in [3, 4, 5]:
        return 1 if numeric else 'spring'
    if x in [6, 7, 8]:
        return 2 if numeric else 'summer'
    if x in [9, 10, 11]:
        return  3 if numeric else 'fall'
      
# all lower-letter columns are the derived ones.
final_df['season'] = final_df.month.apply(lambda x: make_season(x))
final_df['numeric_season'] = final_df.month.apply(lambda x: make_season(x, numeric=True))
```

## Statistical Correlation
Our sub-questions:

- Is there a correlation between certain crimes and temperatures? **-0.0123**

- Is there a correlation between different weather conditions and crime in general? **only categorical, hence no numerical correlation performed**

- Is there a correlation between crime and the different seasons? **0.0043**

 


```python
corr_global_temp = final_df.Global_desc.corr(final_df['temp'], method='spearman')
print(corr_global_temp)
```

```python
corr_global_season = final_df.Global_desc.corr(final_df['numeric_season'],method='spearman')
print(corr_global_season)
```
